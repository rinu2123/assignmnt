```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\rinuf\\Downloads\\data.csv")
df
```

```python
len(df)
```

```python
no_val=df[df.isnull().any(axis=1)] # shows dataframe with rows and columns having null values
no_val
```

```python
#First Column
```

```python
j=1   # variable representing column number
i=0   # Row number
for i in range(len(df)+1):  #Iterates the row till the end ie.len of the dataframe
    x=df.iloc[[i,j]]  # Value at specified row and column number 
    y=df.iloc[i+1,j]   # Value just one row down of x
    if y.isnull()==True:
        break
    elif abs(x-y)<=1:
        df.y=df.x
        j=j+1
    else:
        print("No result") 
```

```python
fig=plt.figure()
plt.plot(df.x)
plt.grid(axis='y',linewidth=1)
plt.title("Curve for maxO3")
plt.xlabel("Observed values and Imputed values")
plt.ylabel()
plt.legend()
```
